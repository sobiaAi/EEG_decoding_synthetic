```python
import mne
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

mne.set_log_level('WARNING')
print("MNE verson :", mne.__version__ )
```

```python
from mne.datasets import sample

data_path=sample.data_path()
raw_fname=data_path/'MEG'/'sample'/'sample_audvis_raw.fif'
raw=mne.io.read_raw_fif(raw_fname, preload=True)
print("//data overview")
print(f"duration : {raw.times[-1]:.1f} sec")
print(f"sample frequency : {raw.info['sfreq']} hz")
print(f"no. of channels: {len(raw.ch_names)}")
print(f"channel types: {set(mne.channel_type(raw.info,i) for i in range(len(raw.ch_names)))}")
print(f"\ninitial 5 channels: {raw.ch_names[:5]}")
```

```python
raw_eeg= raw.copy().pick(picks='eeg')
print(f"channels picked: {len(raw_eeg.ch_names)}")
print(f"channel name sample: {raw_eeg.ch_names[:5]}")
raw_eeg.set_eeg_reference('average', projection=True)
raw_eeg.apply_proj()
```

```python
print("bad channels marked in data", raw_eeg.info['bads'])
raw_eeg.interpolate_bads(reset_bads=True)
print("after interpolation, bad channels", raw_eeg.info['bads'])
print("total channels now are", len(raw_eeg.ch_names))
channel_data=raw_eeg.get_data()
channel_var=np.var(channel_data, axis=1)
mean_var=np.mean(channel_var)
stdev_var=np.std(channel_var)
print(f"mean variance across channels: {mean_var:.2e}")
print(f"std deviation of variance: {stdev_var:.2e}")
sus_idx=np.where(np.abs(channel_var-mean_var)>3*stdev_var)[0]
sus_names=[raw_eeg.ch_names[i] for i in sus_idx]
print(f"channels >3 std from mean_var {sus_names}")
```

```python
raw_filtered=raw_eeg.copy().filter(l_freq=1.0, h_freq=40.0)
print("bandpass filter from 1-40 Hz")
print(f"original highpass filter {raw_eeg.info['highpass']} Hz")
print(f"original lowpass filter {raw_eeg.info['lowpass']} Hz")
print("my filter - lowcut: 1.0 hz and highcut: 40.0 hz")
```

```python
# find events
events=mne.find_events(raw, stim_channel='STI 014', verbose= False)
print(f"total events found {len(events)}")
print(f"\n what an event looks like\n  {events[:3]}")
print("\n columns: [sample_no., pre_eid, curr_eid] ")
event_id= {
    'auditory/left':1 ,
    'auditory/right':2 ,
    'visual/left':3 ,
    'visual/right':4
}

print("trials per condition")
for name, eid in event_id.items():
    count= np.sum(events[:,2]==eid)
    print(f"{name} : {count} trials")

epochs=mne.Epochs(
    raw_filtered,
    events,
    event_id=event_id,
    tmin=-0.2,
    tmax=0.5,
    baseline=(-0.2,0),
    preload= True,
    reject=dict(eeg=150e-6)
)
print(f"epochs after artifact rejection {len(epochs)}")
print(f"shape :{epochs.get_data().shape}")
print(f"time runs from {epochs.tmin*1000:.0f} ms to {epochs.tmax*1000:.0f} ms")
```

```python
erp_aud=epochs['auditory/left'].average()
erp_vis=epochs['visual/left'].average()

print(f"shape: {erp_aud.data.shape} (channels x timepoints)")
print(f"time from {erp_aud.times[0]*1000:.0f}ms to {erp_aud.times[-1]*1000:.0f}ms")
peak_idx=np.argmax(np.abs(erp_aud.data).mean(axis=0))
peak_time=erp_aud.times[peak_idx]*1000
print(f"peak mean apmlitude at: {peak_time:.1f}ms (auditory/left)")
```

```python
psd_aud=epochs['auditory/left'].compute_psd(method='welch', fmin=1, fmax=40)
psd_vis=epochs['visual/left'].compute_psd(method='welch', fmin=1, fmax=40)

freqs=psd_aud.freqs
power_aud=psd_aud.get_data().mean(axis=(0,1))
power_vis=psd_vis.get_data().mean(axis=(0,1))
print(f"power spectral sensity computed from {freqs[0]:.1f}Hz to {freqs[-1]:.1f}Hz")
print(f"number of freqs: {len(freqs)}")
alpha_mask= (freqs>=8) & (freqs<=13)
print(f"\nAlpha band (8-13 Hz) mean power:")
print(f"  Auditory: {power_aud[alpha_mask].mean():.4e}")
print(f"  Visual:   {power_vis[alpha_mask].mean():.4e}")
print(f"\nBeta band (13-30 Hz) mean power:")
print(f"  Auditory: {power_aud[(freqs>=13)&(freqs<=30)].mean():.4e}")
print(f"  Visual:   {power_vis[(freqs>=13)&(freqs<=30)].mean():.4e}")
```

```python
tmin_stat, tmax_stat=0.1,0.3
time_mask=(epochs.times>=tmin_stat) & (epochs.times<=tmax_stat)
aud_data = epochs['auditory/left'].get_data(picks='eeg')[:, :, time_mask].mean(axis=(1,2))
vis_data = epochs['visual/left'].get_data(picks='eeg')[:, :, time_mask].mean(axis=(1,2))

t_stat, p_value = stats.ttest_ind(aud_data, vis_data)
print(f"auditory mean amplitude: {aud_data.mean()*1e6:.3f} µV")
print(f"visual mean amplitude: {vis_data.mean()*1e6:.3f} µV")
print(f"t = {t_stat:.3f},  p = {p_value:.4f}")
print("Significant difference." if p_value < 0.05 else "No significant difference.")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('EEG Data Cleaning & Analysis — MNE Sample Dataset', 
             fontsize=14, fontweight='bold')

#Raw vs Filtered signal...
ax1 = axes[0, 0]
ch_idx = 0
t_slice = slice(0, int(5 * raw_eeg.info['sfreq']))
raw_data = raw_eeg.get_data()[ch_idx, t_slice]
filt_data = raw_filtered.get_data()[ch_idx, t_slice]
times_raw = raw_eeg.times[t_slice]

ax1.plot(times_raw, raw_data * 1e6, alpha=0.5, label='Raw', color='gray')
ax1.plot(times_raw, filt_data * 1e6, label='Filtered (1-40Hz)', color='steelblue')
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Amplitude (µV)')
ax1.set_title('Raw vs Filtered Signal')
ax1.legend()

#Plot 2: ERP comparison....
ax2 = axes[0, 1]
ch_idx2 = 0
ax2.plot(erp_aud.times * 1000, erp_aud.data[ch_idx2] * 1e6,
         label='Auditory', color='steelblue')
ax2.plot(erp_vis.times * 1000, erp_vis.data[ch_idx2] * 1e6,
         label='Visual', color='coral')
ax2.axvline(0, color='black', linestyle='--', alpha=0.5, label='Stimulus onset')
ax2.axhline(0, color='black', alpha=0.3)
ax2.set_xlabel('Time (ms)')
ax2.set_ylabel('Amplitude (µV)')
ax2.set_title('ERP: Auditory vs Visual')
ax2.legend()

#Plot 3: Power Spectral Density...
ax3 = axes[1, 0]
ax3.semilogy(freqs, power_aud, label='Auditory', color='steelblue')
ax3.semilogy(freqs, power_vis, label='Visual', color='coral')
ax3.axvspan(8, 13, alpha=0.15, color='green', label='Alpha band')
ax3.set_xlabel('Frequency (Hz)')
ax3.set_ylabel('Power (µV²/Hz)')
ax3.set_title('Power Spectral Density')
ax3.legend()

#Plot 4: Trial amplitude distributions...
ax4 = axes[1, 1]
ax4.hist(aud_data * 1e6, bins=20, alpha=0.6, label='Auditory', color='steelblue')
ax4.hist(vis_data * 1e6, bins=20, alpha=0.6, label='Visual', color='coral')
ax4.set_xlabel('Mean Amplitude 100-300ms (µV)')
ax4.set_ylabel('Trial count')
ax4.set_title(f'Amplitude Distribution (p = {p_value:.4f})')
ax4.legend()

plt.tight_layout()
plt.savefig('eeg_analysis_results.png', dpi=150, bbox_inches='tight')
plt.show()
print("Figure saved.")
```
